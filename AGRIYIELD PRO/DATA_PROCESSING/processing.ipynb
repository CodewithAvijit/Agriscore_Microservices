```python
import pandas as pd
import joblib
import numpy as np
import os
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
path=os.path.join('c:\\Users\\Avijit\\Desktop\\AgriAssure\\AGRIYIELD PRO\\DATASET',"crop_yield.csv")
```

```python
data=pd.read_csv(path)
data.head()
```

```python
crop_encoder=LabelEncoder()
season_encoder=LabelEncoder()
scaler=StandardScaler()
```

```python
data['Crop']=crop_encoder.fit_transform(data['Crop'])
data['Season']=season_encoder.fit_transform(data['Season'])
```

```python
joblib.dump(crop_encoder,'c:\\Users\\Avijit\\Desktop\\AgriAssure\\AGRIYIELD PRO\\ENCODER-DECODER\\crop.pkl')
joblib.dump(season_encoder,'c:\\Users\\Avijit\\Desktop\\AgriAssure\\AGRIYIELD PRO\\ENCODER-DECODER\\season.pkl')
```

```python
data.iloc[:,:-1]=scaler.fit_transform(data.iloc[:,:-1])
```

```python
joblib.dump(scaler,"c:\\Users\\Avijit\\Desktop\\AgriAssure\\AGRIYIELD PRO\\ENCODER-DECODER\\scaler.pkl")
```

```python
data['Yield'] = data['Yield'] * 1000
```

```python
Q1 = data['Yield'].quantile(0.25)
Q3 = data['Yield'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Yield'] < lower_bound) | (data['Yield'] > upper_bound)]
num_outliers = outliers.shape[0]

print(f"Number of outliers in 'Yield': {num_outliers}")
data = data[(data['Yield'] >= lower_bound) & (data['Yield'] <= upper_bound)]
print(f"Remaining data after outlier removal: {data.shape[0]} rows")
```

```python
percent = (num_outliers / data.shape[0]) * 100
print(f"Outliers percentage in 'Yield': {percent:.2f}%")
```

Crop: The name of the crop cultivated.
Crop_Year: The year in which the crop was grown.

Area: The total land area (in hectares) under cultivation for the specific crop.
