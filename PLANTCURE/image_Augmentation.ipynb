```python
import os
import shutil

# Root folder containing multiple subfolders with images
root_folder = r"C:\Users\Avijit\Desktop\AgriAssure\PLANTCURE\DATASET\archive\FSS-1000"

# Destination folder where all images will be copied
destination_folder = r"C:\Users\Avijit\Desktop\AgriAssure\PLANTCURE\DATASET\PLANTorSOMETHING\OTHERS"

# Make sure destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Supported image extensions
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.webp')

# Walk through all subdirectories in the root folder
for subdir, _, files in os.walk(root_folder):
    for file in files:
        if file.lower().endswith(image_extensions):
            src_file = os.path.join(subdir, file)
            # Use subfolder name to avoid overwriting files with same name
            folder_name = os.path.basename(subdir)
            unique_name = folder_name + "_" + file
            
            dst_file = os.path.join(destination_folder, unique_name)

            shutil.copy(src_file, dst_file)

print("âœ… All images copied from subfolders to target folder successfully.")
```

```python
import os
import shutil
import random

plant_dir = r"C:\Users\Avijit\Desktop\AgriAssure\PLANTCURE\DATASET\Healthcheck\HEALTHY"
others_dir = r"C:\Users\Avijit\Desktop\AgriAssure\PLANTCURE\DATASET\Healthcheck\UNHEALTHY"

# Get all image filenames
plant_images = [f for f in os.listdir(plant_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
others_images = [f for f in os.listdir(others_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

plant_count = len(plant_images)
others_count = len(others_images)

print(f"PLANT images: {plant_count}")
print(f"OTHERS images: {others_count}")

# Calculate how many more images are needed
difference = others_count - plant_count

if difference <= 0:
    print("âœ… No need to copy, PLANT already has enough images.")
else:
    print(f"ðŸ“ˆ Need to copy {difference} more images to PLANT folder...")

    for i in range(difference):
        # Randomly select an image from existing PLANT images
        img_to_copy = random.choice(plant_images)
        src_path = os.path.join(plant_dir, img_to_copy)

        # Create a unique name for the copied file
        name, ext = os.path.splitext(img_to_copy)
        new_filename = f"{name}_copy{i}{ext}"
        dst_path = os.path.join(plant_dir, new_filename)

        shutil.copy(src_path, dst_path)

    print("âœ… Done! PLANT folder now has the same number of images as OTHERS.")
```

```python
import os
import shutil
from pathlib import Path

def copy_class_folders_only(source_folders, destination_folder):
    os.makedirs(destination_folder, exist_ok=True)

    for source in source_folders:
        source_path = Path(source)
        if not source_path.exists():
            print(f"âŒ Source folder not found: {source}")
            continue

        # Get only immediate subdirectories (classes)
        for class_folder in source_path.iterdir():
            if class_folder.is_dir():
                destination_class_folder = Path(destination_folder) / class_folder.name
                os.makedirs(destination_class_folder, exist_ok=True)
                print(f"ðŸ“ Created empty folder: {destination_class_folder}")

# âœ… Example usage:
source_folders = [
    r"C:\Users\Avijit\Desktop\archive\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\train"
]

destination_folder = r"C:\Users\Avijit\Desktop\AgriAssure\PLANTCURE\DATASET\PlantVillage"

copy_class_folders_only(source_folders, destination_folder)
```

```python
import zipfile
from pathlib import Path

def unzip_file(zip_path, extract_to):
    zip_path = Path(zip_path)
    extract_to = Path(extract_to)

    if not zip_path.exists() or not zip_path.suffix == ".zip":
        print(f"âŒ Invalid zip file: {zip_path}")
        return

    # Create destination folder if it doesn't exist
    extract_to.mkdir(parents=True, exist_ok=True)

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
        print(f"âœ… Extracted '{zip_path.name}' to '{extract_to}'")

# âœ… Example usage:
zip_file_path = r"C:\Users\Avijit\Desktop\DEEP_LEARNING\datasets\archive.zip"
destination_folder = r"C:\Users\Avijit\Desktop\archive"

unzip_file(zip_file_path, destination_folder)
```
