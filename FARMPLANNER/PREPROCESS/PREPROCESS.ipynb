```python
import pandas as pd
import numpy as np
import joblib as jb
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
data = pd.read_csv("C:/Users/Avijit/Desktop/AgriAssure/CROP_RECOMMENDATION/DATASET/Crop_recommendation.csv")
```

```python
label_en=LabelEncoder()
```

```python
data['label']=label_en.fit_transform(data['label'])
```

```python
data.iloc[:,:3]=data.iloc[:,:3].astype(float)
```

```python
scalerx=StandardScaler()
```

```python
data.iloc[:,:-1]=scalerx.fit_transform(data.iloc[:,:-1])
```

```python
jb.dump(scalerx, r"C:\Users\Avijit\Desktop\AgriAssure\CROP_RECOMMENDATION\ENCODER-DECODER\scalerx.pkl")
jb.dump(label_en, r"C:\Users\Avijit\Desktop\AgriAssure\CROP_RECOMMENDATION\ENCODER-DECODER\label_en.pkl")
```

```python
data.to_csv(r"C:\Users\Avijit\Desktop\AgriAssure\CROP_RECOMMENDATION\PROCESS_DATASET/crop_recommendation_processed.csv",index=False)
```
